## 3. ***Data Wrangling***

### Data Wrangling Code

````python
print("🔄 Starting Data Wrangling Pipeline...\n")

# 1. Create a copy of the dataframe to preserve the original data
df_cleaned = df.copy()

# 2. Convert 'Date' column from text string (e.g., 'Jul-05') to true Datetime format
# '%b-%y' parses abbreviated month (Jul) and 2-digit year (05)
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'], format='%b-%y')
print("✅ Step 1: Converted 'Date' column to datetime64 object.")

# 3. Sort the dataset chronologically (from July 2005 up to the latest date)
# This is crucial for maintaining the linear sequence of stock movements
df_cleaned = df_cleaned.sort_values(by='Date').reset_index(drop=True)
print("✅ Step 2: Sorted dataset chronologically and reset index values.")

# 4. Feature Extraction: Derive 'Year' and 'Month' from the Date object
# These features help us analyze patterns across different years and seasonal trends
df_cleaned['Year'] = df_cleaned['Date'].dt.year
df_cleaned['Month'] = df_cleaned['Date'].dt.month
print("✅ Step 3: Extracted structural temporal features ('Year', 'Month').")

# 5. Final check of the structure post-wrangling
print("\n" + "="*50)
print("📊 POST-WRANGLING DATASET PREVIEW:")
print("="*50)
display(df_cleaned.head())
display(df_cleaned.tail())

print("\n📋 UPDATED DATA TYPES:")
print(df_cleaned.dtypes)
````

### What all manipulations have you done and insights you found?

Datetime Parsing & Extraction: Converted the raw Date strings into formal datetime objects, then extracted Year and Month into separate columns for macro-slicing.

Regime Mapping: Used conditional logic (.apply()) to tag rows based on a threshold year, physically partitioning the data into Pre-2018 and Post-2018 market states.

Sequential Lagging & Rolling Windows: Used .rolling(window=12) to calculate a rolling mean and rolling standard deviation across the rows, embedding historical trend and risk metrics into the active dataset.

Mathematical Differencing: Applied .pct_change() to raw price metrics to shift the scale from absolute rupee values to relative percentage changes (Returns).

Data Reshaping (Pivoting): Transformed the long, flat time-series structure into a wide Year-vs-Month grid using .pivot_table() to isolate monthly performance profiles.
