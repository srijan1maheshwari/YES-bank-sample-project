```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Hand Tracking Particles</title>
    <style>
        body { margin: 0; overflow: hidden; background: #000; font-family: sans-serif; }
        #ui { position: absolute; top: 20px; left: 20px; color: white; z-index: 10; pointer-events: none; }
        .hint { font-size: 0.8em; opacity: 0.7; }
        video { transform: scaleX(-1); position: absolute; width: 200px; height: auto; bottom: 20px; right: 20px; border-radius: 10px; border: 2px solid #444; }
        canvas { display: block; }
    </style>
</head>
<body>

<div id="ui">
    <h1>Particle Alchemist</h1>
    <p>Pinch fingers to expand | Move hand to steer</p>
    <div id="status">Shape: Sphere</div>
</div>

<video id="input_video"></video>

<script>
/**
 * CONFIGURATION & SHAPE MATH
 */
const PARTICLE_COUNT = 5000;
const transitionSpeed = 0.05;
let currentShape = 'sphere';

const shapes = {
    sphere: (i) => {
        const phi = Math.acos(-1 + (2 * i) / PARTICLE_COUNT);
        const theta = Math.sqrt(PARTICLE_COUNT * Math.PI) * phi;
        return {
            x: Math.cos(theta) * Math.sin(phi) * 2,
            y: Math.sin(theta) * Math.sin(phi) * 2,
            z: Math.cos(phi) * 2
        };
    },
    heart: (i) => {
        const t = (i / PARTICLE_COUNT) * Math.PI * 2;
        const x = 16 * Math.pow(Math.sin(t), 3);
        const y = 13 * Math.cos(t) - 5 * Math.cos(2 * t) - 2 * Math.cos(3 * t) - Math.cos(4 * t);
        return { x: x * 0.15, y: y * 0.15, z: (Math.random() - 0.5) * 0.5 };
    },
    saturn: (i) => {
        if (i < PARTICLE_COUNT * 0.6) { // The Planet
            return shapes.sphere(i);
        } else { // The Rings
            const angle = Math.random() * Math.PI * 2;
            const radius = 3 + Math.random() * 1.5;
            return {
                x: Math.cos(angle) * radius,
                y: (Math.random() - 0.5) * 0.2,
                z: Math.sin(angle) * radius
            };
        }
    },
    flower: (i) => {
        const t = (i / PARTICLE_COUNT) * Math.PI * 20;
        const r = 2 * Math.cos(5/3 * t); // Rose curve
        return {
            x: r * Math.cos(t),
            y: r * Math.sin(t),
            z: (i / PARTICLE_COUNT) * 2
        };
    }
};

/**
 * THREE.JS SETUP
 */
const scene = new THREE.Scene();
const camera = new THREE.PerspectiveCamera(75, window.innerWidth / window.innerHeight, 0.1, 1000);
const renderer = new THREE.WebGLRenderer({ antialias: true });
renderer.setSize(window.innerWidth, window.innerHeight);
document.body.appendChild(renderer.domElement);

// Particle Geometry
const geometry = new THREE.BufferGeometry();
const positions = new Float32Array(PARTICLE_COUNT * 3);
const targetPositions = new Float32Array(PARTICLE_COUNT * 3);
const colors = new Float32Array(PARTICLE_COUNT * 3);

for (let i = 0; i < PARTICLE_COUNT; i++) {
    const pos = shapes.sphere(i);
    positions[i * 3] = pos.x;
    positions[i * 3 + 1] = pos.y;
    positions[i * 3 + 2] = pos.z;
    targetPositions[i * 3] = pos.x;
    targetPositions[i * 3 + 1] = pos.y;
    targetPositions[i * 3 + 2] = pos.z;

    colors[i * 3] = Math.random();
    colors[i * 3 + 1] = Math.random();
    colors[i * 3 + 2] = 1.0;
}

geometry.setAttribute('position', new THREE.BufferAttribute(positions, 3));
geometry.setAttribute('color', new THREE.BufferAttribute(colors, 3));

const material = new THREE.PointsMaterial({
    size: 0.05,
    vertexColors: true,
    transparent: true,
    blending: THREE.AdditiveBlending
});

const particleSystem = new THREE.Points(geometry, material);
scene.add(particleSystem);

camera.position.z = 10;

/**
 * HAND TRACKING LOGIC
 */
const videoElement = document.getElementById('input_video');
let handX = 0, handY = 0, pinchDist = 1;

function onResults(results) {
    if (results.multiHandLandmarks && results.multiHandLandmarks.length > 0) {
        const hand = results.multiHandLandmarks[0];

        // Map hand coordinates (-0.5 to 0.5) to Three.js space
        handX = (hand[8].x - 0.5) * -20;
        handY = (hand[8].y - 0.5) * -15;

        // Detect Pinch (Thumb tip to Index tip)
        const dx = hand[4].x - hand[8].x;
        const dy = hand[4].y - hand[8].y;
        pinchDist = Math.sqrt(dx*dx + dy*dy) * 10;

        // Change shape based on palm orientation or simple timer for demo
        // For this demo: Switch shape if hand is very high
        if (hand[8].y < 0.2) updateShape('heart');
        else if (hand[8].x < 0.2) updateShape('saturn');
        else if (hand[8].x > 0.8) updateShape('flower');
        else if (hand[8].y > 0.8) updateShape('sphere');
    }
}

function updateShape(key) {
    if (currentShape === key) return;
    currentShape = key;
    document.getElementById('status').innerText = "Shape: " + key.toUpperCase();
    for (let i = 0; i < PARTICLE_COUNT; i++) {
        const p = shapes[key](i);
        targetPositions[i * 3] = p.x;
        targetPositions[i * 3 + 1] = p.y;
        targetPositions[i * 3 + 2] = p.z;
    }
}

const hands = new Hands({
    locateFile: (file) => `https://cdn.jsdelivr.net/npm/@mediapipe/hands/${file}`
});
hands.setOptions({
    maxNumHands: 1,
    modelComplexity: 1,
    minDetectionConfidence: 0.5,
    minTrackingConfidence: 0.5
});
hands.onResults(onResults);

const cameraUtils = new Camera(videoElement, {
    onFrame: async () => { await hands.send({image: videoElement}); },
    width: 640, height: 480
});
cameraUtils.start();

/**
 * ANIMATION LOOP
 */
function animate() {
    requestAnimationFrame(animate);

    const posAttr = geometry.attributes.position;

    for (let i = 0; i < PARTICLE_COUNT * 3; i++) {
        // Smoothly interpolate to target shape
        posAttr.array[i] += (targetPositions[i] - posAttr.array[i]) * transitionSpeed;
    }

    // Interactive transformations
    particleSystem.position.x += (handX - particleSystem.position.x) * 0.1;
    particleSystem.position.y += (handY - particleSystem.position.y) * 0.1;

    // Scale based on pinch
    const targetScale = 0.5 + pinchDist;
    particleSystem.scale.lerp(new THREE.Vector3(targetScale, targetScale, targetScale), 0.1);

    // Constant rotation
    particleSystem.rotation.y += 0.01;
    particleSystem.rotation.x += 0.005;

    // Dynamic Color Shift
    const time = Date.now() * 0.001;
    material.color.setHSL((time * 0.1) % 1, 0.7, 0.6);

    posAttr.needsUpdate = true;
    renderer.render(scene, camera);
}

animate();

// Handle Resize
window.addEventListener('resize', () => {
    camera.aspect = window.innerWidth / window.innerHeight;
    camera.updateProjectionMatrix();
    renderer.setSize(window.innerWidth, window.innerHeight);
});
</script>
</body>
</html>
```
