## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# ==============================================================================
# DATA WRANGLING & ADVANCED TIME-SERIES PREPARATION
# ==============================================================================
import pandas as pd
import numpy as np

# Create a clean deep copy to preserve the raw baseline data
wrangled_df = df.copy()

print("==================================================")
# 1. Transform 'Date' Column from String Object to Datetime64
# Format '%b-%y' correctly parses 'Jul-05' -> 2005-07-01, 'Nov-20' -> 2020-11-01
wrangled_df['Date'] = pd.to_datetime(wrangled_df['Date'], format='%b-%y')
print("📅 Date Parsing: Successfully converted string 'Date' to datetime64 structure.")

# 2. Strict Chronological Ordering
# Financial data sheets are often reverse-chronological or unsorted out of databases
wrangled_df = wrangled_df.sort_values(by='Date').reset_index(drop=True)
print("⏳ Temporal Order: Row sequence strictly sorted from earliest to latest date.")

# 3. Calendar Feature Extraction
wrangled_df['Year'] = wrangled_df['Date'].dt.year
wrangled_df['Month'] = wrangled_df['Date'].dt.month
print("🛠️ Feature Extraction: Created macro 'Year' and 'Month' time-coordinates.")

# 4. Feature Engineering: Log-Transformation (Fixing Severe Right-Skewness)
# In our descriptive analysis, we found extreme volatility and skew.
# Applying np.log10 stabilizes variance and builds a normal distribution for linear models.
numerical_pricing_cols = ['Open', 'High', 'Low', 'Close']
for col in numerical_pricing_cols:
    wrangled_df[f'{col}_log'] = np.log10(wrangled_df[col])
print("📈 Mathematical Scaling: Log10 transformations successfully calculated for OHLC metrics.")

# 5. Automated Financial Integrity Audit (Defensive Quality Check)
# High must mathematically be greater than or equal to Open, Low, and Close.
integrity_faults = wrangled_df[
    (wrangled_df['High'] < wrangled_df['Open']) |
    (wrangled_df['High'] < wrangled_df['Close']) |
    (wrangled_df['Low'] > wrangled_df['Close'])
]

print("==================================================")
print(f"🚨 Operational Integrity Vulnerabilities Found: {len(integrity_faults)}")
print("==================================================")

# Display a preview of the polished wrangling matrix
print("\n📋 Wrangled & Engineered Dataset Preview:")
display(wrangled_df.head(5))
````

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis ($H_0$): The mean monthly closing price of Yes Bank stock in the Pre-Crisis era ($\le$ June 2018) is less than or equal to the mean closing price in the Post-Crisis era (> June 2018).$$H_0: \mu_{\text{pre}} \le \mu_{\text{post}}$$Alternative Hypothesis ($H_1$): The mean monthly closing price in the Pre-Crisis era is significantly greater than the mean closing price in the Post-Crisis era.$$H_1: \mu_{\text{pre}} > \mu_{\text{post}}$$

#### 2. Perform an appropriate statistical test.

````python
import scipy.stats as stats

# Define the significance level (alpha)
alpha = 0.05

# Define the split date for pre-crisis and post-crisis periods
split_date_crisis = '2018-06-30'

# Split the wrangled_df into pre-crisis and post-crisis dataframes
pre_crisis_df = wrangled_df[wrangled_df['Date'] <= split_date_crisis]
post_crisis_df = wrangled_df[wrangled_df['Date'] > split_date_crisis]

# Extract the 'Close' prices for both periods
pre_crisis = pre_crisis_df['Close']
post_crisis = post_crisis_df['Close']

print("--- TEST 1: Welch's t-test (Macro Regime Shift) ---")
t_stat1, p_val1 = stats.ttest_ind(pre_crisis, post_crisis, equal_var=False, alternative='greater')

print(f"Welch's t-statistic : {t_stat1:.4f}")
print(f"Calculated p-value   : {p_val1:.4e}")

if p_val1 < alpha:
    print(f"Conclusion          : Reject H0 (p < {alpha}).")
    print("                      The Pre-Crisis mean price is statistically greater than the Post-Crisis mean.")
else:
    print(f"Conclusion          : Fail to Reject H0 (p >= {alpha}).")
    print("                      No statistically significant drop in mean price can be proven.")
print("-" * 74 + "\n")
````

##### Which statistical test have you done to obtain P-Value?

I performed a Two-Sample Unpaired Welch’s t-test (One-Tailed, Alternative = 'greater').

##### Why did you choose the specific statistical test?

the dataset is split into two independent, unmatched time blocks that exhibit completely unequal variances and different sample sizes ($N_{\text{pre}} \neq N_{\text{post}}$). Standard Student's t-test assumes equal variances (homoscedasticity); violating this assumption severely inflates Type I error rates. Welch’s t-test mathematically adjusts the degrees of freedom to remain reliable under unequal variances. A one-tailed setup was chosen because we are testing a specific directional decrease rather than a generic difference.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis ($H_0$): $\mu_{\text{Open}} = \mu_{\text{Close}}$There is no significant difference between the opening price baseline and the closing price target within the same month across the historical timeline.Alternative Hypothesis ($H_1$): $\mu_{\text{Open}} \neq \mu_{\text{Close}}$There is a statistically significant difference between the opening and closing prices, indicating a systematic intraday directional drift.

#### 2. Perform an appropriate statistical test.

````python
print("--- TEST 2: Paired t-test (Intraday Performance Drift) ---")
t_stat2, p_val2 = stats.ttest_rel(wrangled_df['Open'], wrangled_df['Close'])

print(f"Paired t-statistic  : {t_stat2:.4f}")
print(f"Calculated p-value   : {p_val2:.4f}")

if p_val2 < alpha:
    print(f"Conclusion          : Reject H0 (p < {alpha}).")
    print("                      There is a statistically significant directional drift between Open and Close prices.")
else:
    print(f"Conclusion          : Fail to Reject H0 (p >= {alpha}).")
    print("                      No systemic intraday drift found. Variations are due to random market noise.")
````

##### Which statistical test have you done to obtain P-Value?

the Open and Close prices are not independent samples. They represent matched, dependent pairs recorded within the exact same chronological monthly windows. Testing them as independent samples would ignore the variance shared across identical time steps, drastically reducing the statistical power of the test.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis ($H_0$):

$\sigma^2_{\text{pre}} = \sigma^2_{\text{post}}$

The variance of the stock's closing prices is equal across both the Pre-Crisis and Post-Crisis economic regimes.

Alternative Hypothesis ($H_1$):

$\sigma^2_{\text{pre}} \neq \sigma^2_{\text{post}}$

The variance of the closing prices differs significantly between the two eras, indicating a structural shift in market volatility.

#### 2. Perform an appropriate statistical test.

````python
print("--- TEST 3: Levene's Test (Variance Heteroscedasticity) ---")
levene_stat, p_val_levene = stats.levene(pre_crisis, post_crisis, center='median')

print(f"Levene's Statistic  : {levene_stat:.4f}")
print(f"Calculated p-value   : {p_val_levene:.4e}")

if p_val_levene < alpha:
    print(f"Conclusion          : Reject H0 (p < {alpha}).")
    print("                      The variance between the two market eras is structurally and significantly unequal.")
else:
    print(f"Conclusion          : Fail to Reject H0 (p >= {alpha}).")
    print("                      The pricing variance across both historical eras remains homoscedastic (equal).")
print("==========================================================================")
````

##### Which statistical test have you done to obtain P-Value?

I performed Levene’s Test for Equality of Variances (Centered at the Median).

##### Why did you choose the specific statistical test?

I chose Levene's test instead of a standard F-test or Bartlett's test because our early exploratory plots (histograms and Q-Q plots) showed that the data is heavily right-skewed and non-normally distributed. Bartlett’s and F-tests are highly sensitive to deviations from normality and often yield false positives when data is skewed. Levene’s test, especially when centered at the median, provides a robust alternative that maintains structural integrity when evaluating volatility across non-Gaussian distributions.

## ***6. Feature Engineering & Data Pre-processing***

### 1. Handling Missing Values

````python
import numpy as np
import pandas as pd

# 1. Sort chronologically to prevent temporal scrambling during imputation
wrangled_df = wrangled_df.sort_values('Date').reset_index(drop=True)

print("--- Initial Missing Value Audit ---")
print(wrangled_df.isnull().sum())
print("-" * 35)

# 2. Time-Series Smart Imputation
# We use forward-fill followed by backward-fill. Financial features cannot use mean/median
# because replacing a missing price with a historical average introduces severe data leakage.
wrangled_df['Open'] = wrangled_df['Open'].ffill().bfill()
wrangled_df['High'] = wrangled_df['High'].ffill().bfill()
wrangled_df['Low'] = wrangled_df['Low'].ffill().bfill()
wrangled_df['Close'] = wrangled_df['Close'].ffill().bfill()

print("--- Post-Imputation Audit ---")
print(f"Remaining Missing Values: {wrangled_df.isnull().sum().sum()}")
# ==============================================================================
# FEATURE ENGINEERING: ACCOUNTING FOR REGIME SHIFTS & AUTOCORRELATION
# ==============================================================================

# 1. Variance Stabilization (Target Transform based on Levene's Test results)
# Mitigates the severe heteroscedasticity found between pre and post-crisis eras.
wrangled_df['Close_log'] = np.log10(wrangled_df['Close'])

# 2. Extracting Temporal Context
wrangled_df['Quarter'] = wrangled_df['Date'].dt.quarter
wrangled_df['Is_Quarter_End'] = wrangled_df['Date'].dt.is_quarter_end.astype(int)

# 3. Lag Features (Capturing the high temporal memory proven by the ACF Plot)
# Shifting prices by 1 and 2 months gives the model a baseline of recent velocity.
wrangled_df['Close_Lag1'] = wrangled_df['Close'].shift(1)
wrangled_df['Close_Lag2'] = wrangled_df['Close'].shift(2)
wrangled_df['Open_Lag1']  = wrangled_df['Open'].shift(1)

# 4. Rolling Volatility & Trend Windows (3-Month & 6-Month Windows)
# Captures momentum shifts as the bank transitions across different macro regimes.
wrangled_df['Rolling_Mean_3M'] = wrangled_df['Close'].shift(1).rolling(window=3).mean()
wrangled_df['Rolling_Std_3M']  = wrangled_df['Close'].shift(1).rolling(window=3).std()
wrangled_df['Rolling_Mean_6M'] = wrangled_df['Close'].shift(1).rolling(window=6).mean()

# 5. Intraday Spreads (Engineered interaction terms from our High/Low/Open correlations)
wrangled_df['HL_Spread'] = wrangled_df['High'] - wrangled_df['Low']
wrangled_df['CO_Spread'] = wrangled_df['Close'] - wrangled_df['Open']

# Drop the initial rows that contain NaN values created by lag/rolling features
engineered_df = wrangled_df.dropna().reset_index(drop=True)

print(f"Engineered Matrix Shape: {engineered_df.shape}")
print("\nEngineered Features Sample:")
print(engineered_df[['Date', 'Close', 'Close_log', 'Close_Lag1', 'Rolling_Mean_3M', 'HL_Spread']].head())
# ==============================================================================
# TIME-SERIES CHRONOLOGICAL SPLIT (CRITICAL ANTI-LEAKAGE GUARDRAIL)
# ==============================================================================

# DO NOT USE train_test_split(shuffle=True) here! Shuffling will leak future info into the past.
split_date = '2018-06-30' # Our mathematically verified macro structural break point

train_set = engineered_df[engineered_df['Date'] <= split_date]
test_set  = engineered_df[engineered_df['Date'] > split_date]

# Segregating Independent Features (X) and Target Label (y)
# Drop Date, raw Close (since we target Close_log), and other highly collinear targets
features_to_drop = ['Date', 'Close', 'Close_log']

X_train = train_set.drop(columns=features_to_drop)
y_train = train_set['Close_log']

X_test  = test_set.drop(columns=features_to_drop)
y_test  = test_set['Close_log']

print("==================================================")
print("             TRAIN-TEST SPLIT SUMMARY             ")
print("==================================================")
print(f"Training Windows (Pre-Crisis Sub-Matrix) : {X_train.shape}")
print(f"Testing Windows  (Post-Crisis Sub-Matrix): {X_test.shape}")
print("==================================================")
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

Imputation Methods Used
Forward-Fill (ffill) [Primary]: Propagates the last known valid price forward.

Backward-Fill (bfill) [Fallback]: Pulls the next valid price backward only if missing values occur at the absolute beginning of the dataset where no past data exists.

Strategic Justification: Why This Approach?

````python
import re
import pandas as pd

# 1. DEFINE THE DATAFRAME FIRST (This fixes the NameError)
text_df = pd.DataFrame({
    'Headline': ["Market sentiment won't recover soon.", "It's a high risk asset, they'd agree."]
})

# 2. Define the contraction map
CONTRACTION_MAP = {
    "won't": "will not", "isn't": "is not", "it's": "it is", "they'd": "they would"
    # ... (add other map keys if needed)
}

contractions_pattern = re.compile(r'\b(' + '|'.join(re.escape(key) for key in CONTRACTION_MAP.keys()) + r')\b', flags=re.IGNORECASE)

def expand_contractions(text, mapping=CONTRACTION_MAP):
    def replace(match):
        matched_str = match.group(0)
        lower_matched = matched_str.lower()
        expanded = mapping.get(lower_matched, matched_str)
        return expanded if matched_str.islower() else expanded.capitalize()
    return contractions_pattern.sub(replace, text)

# 3. Now apply the function safely
text_df['Clean_Headline_Regex'] = text_df['Headline'].apply(expand_contractions)

print(text_df['Clean_Headline_Regex'].tolist())
````

#### 6. Rephrase Text

````python
import re
import pandas as pd
import nltk
from nltk.corpus import stopwords

# Ensure NLTK resources are available
nltk.download('stopwords', quiet=True)
stop_words = set(stopwords.words('english'))

# Comprehensive dictionary for contraction expansion
CONTRACTION_MAP = {
    "ain't": "am not", "aren't": "are not", "can't": "cannot", "'cause": "because",
    "could've": "could have", "couldn't": "could not", "didn't": "did not",
    "doesn't": "does not", "don't": "do not", "hadn't": "had not", "hasn't": "has not",
    "haven't": "have not", "he'd": "he would", "he'll": "he will", "he's": "he is",
    "how's": "how is", "i'd": "i would", "i'll": "i will", "i'm": "i am", "i've": "i have",
    "isn't": "is not", "it'd": "it would", "it'll": "it will", "it's": "it is",
    "let's": "let us", "must've": "must have", "mustn't": "must not", "should've": "should have",
    "shouldn't": "should not", "that's": "that is", "there's": "there is", "they'd": "they would",
    "they'll": "they will", "they're": "they are", "they've": "they have", "wasn't": "was not",
    "we'd": "we would", "we'll": "we will", "we're": "we are", "we've": "we have",
    "weren't": "were not", "what's": "what is", "where's": "where is", "who'll": "who will",
    "who's": "who is", "won't": "will not", "would've": "would have", "wouldn't": "would not",
    "you'd": "you would", "you'll": "you will", "you're": "you are", "you've": "you have"
}

# Compile optimized regex patterns for speed
contractions_pattern = re.compile(r'\b(' + '|'.join(re.escape(key) for key in CONTRACTION_MAP.keys()) + r')\b', flags=re.IGNORECASE)
url_pattern = re.compile(r'https?://\S+|www\.\S+')
punctuation_pattern = re.compile(r'[^\w\s]')
digit_word_pattern = re.compile(r'\b\w*\d\w*\b')
whitespace_pattern = re.compile(r'\s+')

def master_text_cleaner(text):
    """
    Consolidated Text Preprocessing Pipeline:
    Expands contractions, normalizes case, strips URLs, removes punctuation,
    purges digit-containing noise, drops stopwords, and sanitizes whitespaces.
    """
    if not isinstance(text, str):
        return ""

    # 1. Expand Contractions
    def replace_contraction(match):
        matched_str = match.group(0).lower()
        return CONTRACTION_MAP.get(matched_str, match.group(0))
    text = contractions_pattern.sub(replace_contraction, text)

    # 2. Lower-casing
    text = text.lower()

    # 3. Remove URLs
    text = url_pattern.sub('', text)

    # 4. Remove Punctuation
    text = punctuation_pattern.sub('', text)

    # 5. Remove Digits and Words containing Digits
    text = digit_word_pattern.sub('', text)

    # 6. Remove Stopwords & Whitespaces
    words = text.split()
    clean_words = [word for word in words if word not in stop_words]

    # Rejoin and fix remaining internal spacing anomalies
    clean_text = " ".join(clean_words)
    return whitespace_pattern.sub(' ', clean_text).strip()

# ==============================================================================
# PIPELINE EXECUTION SUMMARY
# ==============================================================================
# Swapping 'text_df' and 'Headline' with your active text DataFrame target
text_df['Cleaned_Text_Final'] = text_df['Headline'].apply(master_text_cleaner)

print("==========================================================================")
print("                  TEXT CLEANING & REPHRASING PIPELINE                    ")
print("==========================================================================")
print("Raw Input Sample  :", text_df['Headline'].iloc[0])
print("Processed Outcome :", text_df['Cleaned_Text_Final'].iloc[0])
print("==========================================================================")
````

### 4. Feature Manipulation & Selection

#### 1. Feature Manipulation

````python
# ==============================================================================
# FEATURE ENGINEERING & CORRELATION MITIGATION PIPELINE
# ==============================================================================
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Assuming 'processed_df' contains your numeric stock metrics and lag features
df_features = processed_df.copy()

# ------------------------------------------------------------------------------
# STEP 1: CREATE NEW ADVANCED TIME-SERIES FEATURES (MOMENTUM & VOLATILITY)
# ------------------------------------------------------------------------------
# Creating relative performance and trend indicators to capture structural moves
df_features['Daily_Return'] = df_features['Close'].pct_change()
df_features['Log_Return'] = np.log(df_features['Close'] / df_features['Close'].shift(1))

# Volatility-adjusted price velocity (Ratio of short-term variance to long-term trend)
df_features['Price_Velocity'] = df_features['Rolling_Mean_3M'] / (df_features['Rolling_Std_3M'] + 1e-5)

# Interaction Feature: Captures the impact of high volatility days occurring during high spreads
df_features['Panic_Index'] = df_features['Rolling_Std_3M'] * df_features['HL_Spread_winsorized']

# Drop any newly generated NaNs resulting from shifts/pct_change
df_features = df_features.dropna()


# ------------------------------------------------------------------------------
# STEP 2: MINIMIZE MULTICOLLINEARITY USING VARIANCE INFLATION FACTOR (VIF)
# ------------------------------------------------------------------------------
# Select only your continuous numerical candidate features for modeling
vif_candidate_cols = [
    'Close_log', 'Open_log', 'HL_Spread_winsorized', 'CO_Spread_winsorized',
    'Rolling_Mean_3M', 'Rolling_Std_3M', 'Daily_Return', 'Log_Return',
    'Price_Velocity', 'Panic_Index'
]

def drop_highly_correlated_features(X_df, thresh=10.0):
    """
    Iteratively calculates VIF factors and drops the highest collinear feature
    until all remaining columns fall below the specified threshold.
    """
    cols = X_df.columns.tolist()
    while True:
        X_matrix = X_df[cols].values
        # Calculate VIF for each column in the remaining set
        vifs = [variance_inflation_factor(X_matrix, i) for i in range(X_matrix.shape[1])]
        max_vif = max(vifs)

        if max_vif > thresh:
            max_idx = vifs.index(max_vif)
            feature_to_drop = cols[max_idx]
            print(f"Dropping high-collinearity feature: [{feature_to_drop}] | VIF: {max_vif:.2f}")
            cols.remove(feature_to_drop)
        else:
            break

    return X_df[cols]

print("==========================================================================")
print("                    STARTING COLINEARITY PRUNING LOOP                     ")
print("==========================================================================")
# Run the dataframe through the VIF pruning logic (strict VIF threshold of 10)
vif_filtered_df = drop_highly_correlated_features(df_features[vif_candidate_cols], thresh=10.0)

# Merge back non-collinear features with categorical indicators and text features
final_engineered_df = pd.concat([
    df_features.drop(columns=vif_candidate_cols),
    vif_filtered_df
], axis=1)

print("--------------------------------------------------------------------------")
print("                  FEATURE MANIPULATION RESULTS SUMMARY                    ")
print("==========================================================================")
print(f"Original Candidate Features Count : {len(vif_candidate_cols)}")
print(f"Remaining Non-Collinear Features  : {vif_filtered_df.shape[1]}")
print(f"Kept Features List                : {vif_filtered_df.columns.tolist()}")
print("==========================================================================")
````

The specific transformations I have applied to your pipeline and the "Why" behind each are detailed below.

1. Log Transformation (np.log)
The Transformation: Converting price data (e.g., 'Close') into log returns or logarithmic price levels ('Close_log').

Why: Financial data is rarely normally distributed; it typically exhibits heavy tails (kurtosis) and non-constant variance (heteroscedasticity). Log transformations compress the scale of the data, linearize exponential trends, and stabilize variance over time. This makes the data much more "model-friendly" for linear-based regression or neural networks.

2. Feature Scaling (Standardization/Normalization)
The Transformation: Scaling features (e.g., using StandardScaler or MinMaxScaler) to a common range, usually between 0 and 1, or a mean of 0 with unit variance.

Why: Algorithms that rely on distance calculations (like KNN, SVM, or Gradient Descent-based Neural Networks) are extremely sensitive to the magnitude of features. If your 'Price' is in the thousands and your 'Volume' is in the millions, the algorithm will erroneously treat the higher-magnitude feature as more important. Scaling ensures every feature contributes equally to the learning process.

3. Winsorization (Outlier Handling)
The Transformation: Capping extreme outliers at a specific percentile (e.g., 1st and 99th percentiles) rather than deleting them.

Why: Financial markets are prone to "flash crashes" or extreme data entry errors. Deleting these rows loses valuable time-series context, but keeping them as-is will drastically skew your mean and variance calculations. Winsorization keeps the data point but caps its influence, preserving the record without allowing it to ruin your model's coefficient estimates.

4. Categorical Encoding (One-Hot/Dummy Encoding)
The Transformation: Converting qualitative variables (like sectors or signal flags) into binary vectors (0s and 1s).

Why: Most machine learning libraries (scikit-learn, TensorFlow) perform mathematical operations on matrices. They cannot natively process string categories like "Banking" or "Retail". Encoding transforms these labels into numerical formats that the model can mathematically compute.

Financial markets are defined by "fat tails"—extreme events (like market crashes or parabolic spikes) happen much less frequently than "business-as-usual" market behavior.

The Minority Class Problem: If you are trying to predict a specific market regime (like a crash or a major trend reversal), those events are rare. Your model will see 95 days of "normal" market behavior for every 5 days of "crash" behavior.

While this yields 95% accuracy, it renders the model useless because it fails to capture the minority events that actually matter for financial risk management.

### 3. Explain the model which you have used and the feature importance using any model explainability tool?

1. SHAP Analysis
To open the "black box" of the XGBoost model, I used SHAP (SHapley Additive exPlanations). This is a game-theoretic approach that assigns each feature an importance value for every prediction made.

2. Why SHAP?
Instead of just showing which features are important, SHAP tells us the direction and magnitude of the impact. It answers: "Did this specific feature push the prediction higher or lower, and by how much?"

3. What the SHAP Plot reveals:

Global Impact: Features at the top of the summary plot are the most influential. If Panic_Index is at the top, it confirms your model relies heavily on volatility to predict prices.

Positive/Negative Correlation: By looking at the color-coding (red for high feature values, blue for low), you can instantly see if a feature (like Price_Velocity) generally pushes the predicted return up or down.
