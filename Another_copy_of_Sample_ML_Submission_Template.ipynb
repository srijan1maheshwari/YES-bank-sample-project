## 3. ***Data Wrangling***

### Data Wrangling Code

````python
import pandas as pd
import re

# 1. Load the datasets
metadata_df = pd.read_csv('Zomato Restaurant names and Metadata.csv')
reviews_df = pd.read_csv('Zomato Restaurant reviews.csv')

# --- PART 1: Wrangling Metadata ---
# Clean the Cost column (remove commas and convert to numeric)
metadata_df['Cost'] = metadata_df['Cost'].replace({',': ''}, regex=True)
metadata_df['Cost'] = pd.to_numeric(metadata_df['Cost'], errors='coerce')

# Fill missing Cost values with the median
metadata_df['Cost'] = metadata_df['Cost'].fillna(metadata_df['Cost'].median())

# --- PART 2: Wrangling Reviews ---
# Clean Rating (convert to numeric, dropping non-numeric rows)
reviews_df['Rating'] = pd.to_numeric(reviews_df['Rating'], errors='coerce')
reviews_df.dropna(subset=['Rating'], inplace=True)

# Parse the Metadata column (e.g., "1 Review , 2 Followers")
# This creates two new columns: 'Review_Count' and 'Follower_Count'
reviews_df[['Review_Count', 'Follower_Count']] = reviews_df['Metadata'].str.extract(r'(\d+) Review.*, (\d+) Follower')
reviews_df['Review_Count'] = pd.to_numeric(reviews_df['Review_Count'], errors='coerce').fillna(0)
reviews_df['Follower_Count'] = pd.to_numeric(reviews_df['Follower_Count'], errors='coerce').fillna(0)

# Convert Time to datetime objects
reviews_df['Time'] = pd.to_datetime(reviews_df['Time'], errors='coerce')

# --- PART 3: Merging ---
# Join the two datasets on the restaurant name
# We perform an 'inner' join to keep only restaurants present in both files
merged_df = pd.merge(
    reviews_df,
    metadata_df,
    left_on='Restaurant',
    right_on='Name',
    how='inner'
)

# --- PART 4: Final Hygiene ---
# Remove duplicates that might have been created during the merge
merged_df.drop_duplicates(inplace=True)

# Remove the original columns that we have parsed/replaced
merged_df.drop(columns=['Metadata', 'Name'], inplace=True)

# Display the result
print("Data Wrangling Complete.")
print(f"Final Dataset Shape: {merged_df.shape}")
print(merged_df.head())
````

### What all manipulations have you done and insights you found?

We treated our raw CSV files as "dirty" inputs and systematically cleaned them to ensure the machine can "read" them correctly.

Standardizing Numeric Data: We took columns like Cost and Rating, which were initially loaded as text, and stripped away formatting characters (like commas) or non-numeric noise. By forcing these into float or integer types, we can now perform mathematical calculations like finding the average, minimum, and maximum costs or ratings.

Feature Extraction from Unstructured Text: The most significant manipulation was the Metadata column in your reviews file. It was a "messy" string containing multiple data points ("1 Review, 2 Followers"). We utilized Regular Expressions (regex) to parse that single string into two clean, independent, and numerical columns: Review_Count and Follower_Count.

Temporal Normalization: We converted the Time column into a proper datetime object. This was crucial; it transforms the time from a mere text label into a variable the computer can understand as a progression, allowing you to plot trends over years, months, or even hours.

Relational Merging: We acted as a database engineer by performing an inner join. This combined the reviews and metadata files based on the common Restaurant/Name key. This effectively mapped the "Voice of the Customer" to the "Restaurant Profile," creating a single, enriched dataset.

Hygiene and Noise Reduction: We systematically removed duplicate rows and dropped rows with null ratings. This ensures that when we train a model, it isn't "learning" from junk data or accidentally weighing a single review multiple times.

### 1. Handling Missing Values

````python
import pandas as pd
import re

# Re-load the datasets and perform wrangling steps to ensure merged_df is available
reviews_df = pd.read_csv('/content/Zomato Restaurant reviews.csv')
metadata_df = pd.read_csv('/content/Zomato Restaurant names and Metadata.csv')

# --- PART 1: Wrangling Metadata ---
# Clean the Cost column (remove commas and convert to numeric)
metadata_df['Cost'] = metadata_df['Cost'].replace({',': ''}, regex=True)
metadata_df['Cost'] = pd.to_numeric(metadata_df['Cost'], errors='coerce')

# Fill missing Cost values with the median
metadata_df['Cost'] = metadata_df['Cost'].fillna(metadata_df['Cost'].median())

# --- PART 2: Wrangling Reviews ---
# Clean Rating (convert to numeric, dropping non-numeric rows)
reviews_df['Rating'] = pd.to_numeric(reviews_df['Rating'], errors='coerce')
reviews_df.dropna(subset=['Rating'], inplace=True)

# Parse the Metadata column (e.g., "1 Review , 2 Followers")
# This creates two new columns: 'Review_Count' and 'Follower_Count'
reviews_df[['Review_Count', 'Follower_Count']] = reviews_df['Metadata'].str.extract(r'(\d+) Review.*, (\d+) Follower')
reviews_df['Review_Count'] = pd.to_numeric(reviews_df['Review_Count'], errors='coerce').fillna(0)
reviews_df['Follower_Count'] = pd.to_numeric(reviews_df['Follower_Count'], errors='coerce').fillna(0)

# Convert Time to datetime objects
reviews_df['Time'] = pd.to_datetime(reviews_df['Time'], errors='coerce')

# --- PART 3: Merging ---
# Join the two datasets on the restaurant name
# We perform an 'inner' join to keep only restaurants present in both files
merged_df = pd.merge(
    reviews_df,
    metadata_df,
    left_on='Restaurant',
    right_on='Name',
    how='inner'
)

# --- PART 4: Final Hygiene ---
# Remove duplicates that might have been created during the merge
merged_df.drop_duplicates(inplace=True)

# Remove the original columns that we have parsed/replaced
merged_df.drop(columns=['Metadata', 'Name'], inplace=True)


# First, identify the scale of the problem
print("Missing values before imputation:\n", merged_df.isnull().sum())

# --- Technique A: Median Imputation for Numeric Features ---
# Used for 'Cost' and 'Rating'.
# We use .median() because it is robust against extreme outliers.
merged_df['Cost'] = merged_df['Cost'].fillna(merged_df['Cost'].median())
merged_df['Rating'] = merged_df['Rating'].fillna(merged_df['Rating'].median())

# --- Technique B: Mode (Most Frequent) Imputation for Categorical Features ---
# Used for 'Cuisines' or 'Area' columns.
# We fill with the most common value to maintain the general distribution.
if 'Cuisines' in merged_df.columns:
    most_common_cuisine = merged_df['Cuisines'].mode()[0]
    merged_df['Cuisines'] = merged_df['Cuisines'].fillna(most_common_cuisine)

# --- Technique C: Dropping Rows (The Last Resort) ---
# If a row is missing the critical target variable (e.g., 'Restaurant Name'),
# it is better to drop it than to guess the identity.
merged_df.dropna(subset=['Restaurant'], inplace=True)

print("\nMissing values after imputation:\n", merged_df.isnull().sum())
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

A. Median Imputation (for Numeric Columns like Cost and Rating)
Why I used it: The mean (average) is highly sensitive to outliers—a single "luxury" restaurant with an incredibly high price would skew the mean for the entire dataset. The median represents the 50th percentile, meaning it acts as the "typical" restaurant price or rating. It allows us to fill in gaps without distorting the data with extreme values.

B. Mode Imputation (for Categorical Columns like Cuisines)
Why I used it: Categorical data cannot be "averaged" (you cannot have an average of 'Italian' and 'Chinese'). The mode (the most frequently occurring value) is the most logical substitute. By assigning the most popular cuisine to missing fields, we ensure the variable remains representative of the overall dataset profile without introducing new, non-existent categories.

C. Row Dropping (for Critical Identification Data)
Why I used it: Imputation is essentially an educated guess. If a critical identifier like the Restaurant name is missing, the row loses its utility; it becomes "orphaned" data that cannot be mapped or analyzed correctly. Dropping these rows is the only way to ensure the integrity of the dataset. It is a "cleanliness over quantity" approach—better to lose a few rows than to introduce fake information into your analysis.

### 2. Handling Outliers

````python
# Function to treat outliers using the IQR method and Clipping
def treat_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # We clip the values: anything above/below bounds is set to the bound limit
    # This keeps the data point but removes the extreme distortion
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df

# Apply to the 'Cost' column (as it is continuous and prone to outliers)
merged_df = treat_outliers(merged_df, 'Cost')
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Technique A: Interquartile Range (IQR) Method
What it is: This method calculates the range between the 25th percentile (Q1) and the 75th percentile (Q3). Any data point falling 1.5 times the IQR above Q3 or below Q1 is flagged as an outlier.

Why I used it: Unlike the Z-Score method (which assumes a "Normal" Bell Curve distribution), the IQR method is non-parametric. Restaurant cost data is almost always "Right-Skewed" (many cheap places, a few very expensive ones). The IQR method is mathematically "blind" to the extreme values when calculating the bounds, making it the most reliable way to identify statistical outliers in non-normal datasets.

Technique B: Capping (Winsorization)
What it is: Instead of deleting the outlier rows, we "cap" them. If a restaurant costs 10,000 and our upper bound is 2,000, we change that 10,000 to 2,000.

Why I used it:

Information Preservation: Deleting rows (Trimming) causes data loss, which can bias your sample size and lead to a model that forgets the existence of luxury segments.

Distortion Control: By capping, we keep the record in the dataset (we know this is an "expensive" restaurant), but we prevent its extreme value from mathematically dragging the "Mean" or "Correlation" metrics in your model.

````python
import string
import pandas as pd # Import pandas to use pd.isna

# Function to remove punctuation
def remove_punctuation(text):
    if pd.isna(text): # Check if the value is NaN
        return '' # Return an empty string for NaN values
    # string.punctuation contains: !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~
    return str(text).translate(str.maketrans('', '', string.punctuation))

# Apply the function to the 'review_cleaned' column
merged_df['review_cleaned'] = merged_df['review_cleaned'].apply(remove_punctuation)

print("Punctuation removal complete. Preview:")
print(merged_df[['review_cleaned']].head())
````

````python
import re

def clean_urls_and_digits(text):
    if not isinstance(text, str):
        return ""
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove words containing digits
    text = re.sub(r'\w*\d\w*', '', text)
    return text.strip()

# Apply the cleaning function
merged_df['review_cleaned'] = merged_df['review_cleaned'].apply(clean_urls_and_digits)

print("URLs and digits removed successfully. Preview:")
print(merged_df[['review_cleaned']].head())
````

#### 5. Removing Stopwords & Removing White spaces

````python
import nltk
from nltk.corpus import stopwords

# Ensure stopwords are downloaded
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    # Split text into list of words, filter out stopwords, then join back
    return " ".join([word for word in text.split() if word not in stop_words])

# Apply the function
merged_df['review_cleaned'] = merged_df['review_cleaned'].apply(remove_stopwords)

print("Stopwords removed successfully.")
````

````python
# The split() function naturally handles multiple whitespaces
# and ' '.join() puts them back together with a single space
merged_df['review_cleaned'] = merged_df['review_cleaned'].apply(lambda x: " ".join(x.split()))

print("Whitespace normalized successfully.")
print("Final Cleaned Preview:\n", merged_df['review_cleaned'].head())
````

#### 6. Rephrase Text

````python
import nltk
from nltk.stem import WordNetLemmatizer

# Download necessary data
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    # Split text, lemmatize each word, and join back together
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

# Apply the function
merged_df['review_cleaned'] = merged_df['review_cleaned'].apply(lemmatize_text)

print("Text lemmatization complete.")
print("Preview:\n", merged_df['review_cleaned'].head())
````

#### 8. Text Normalization

````python
from nltk.stem import PorterStemmer

# Initialize the stemmer
stemmer = PorterStemmer()

def stem_text(text):
    # Split text, stem each word, and join back
    return " ".join([stemmer.stem(word) for word in text.split()])

# Apply the function to the already cleaned text
merged_df['review_stemmed'] = merged_df['review_cleaned'].apply(stem_text)

print("Stemming complete. Preview:")
print(merged_df[['review_cleaned', 'review_stemmed']].head())
````

##### Which text vectorization technique have you used and why?

 1. TF-IDF Vectorization

 TF-IDF (Term Frequency-Inverse Document Frequency) is the best choice for restaurant reviews. It doesn't just count words; it weighs them. If a word appears in every review (like "restaurant" or "food"), TF-IDF gives it a low score. If a word appears only in a specific review (like "bland" or "exquisite"), it gives it a high score.

 The mathematical weight is calculated as:

 W_{x,y} = tf_{x,y} \times \log\left(\frac{N}{df_x}\right)

 Where $tf$ is term frequency, $N$ is total reviews, and $df$ is document frequency.

 2. Count Vectorization (Bag of Words)

This is the simplest method. It creates a massive table where columns are every unique word in your entire dataset, and the value is simply how many times that word appears in a specific review (0, 1, 2, etc.).

### 5. Data Transformation

#### Do you think that your data needs to be transformed? If yes, which transformation have you used. Explain Why?

Yes, absolutely. While cleaning (removing noise) and vectorizing (converting text to numbers) are the structural foundations, mathematical transformation is the final step to ensure your model doesn't suffer from bias or skewed data.

Even if you are using a Random Forest (which is robust to feature scales), performing these transformations is best practice. It makes your pipeline "model-agnostic," meaning you can easily switch to a Linear Regression, SVM, or Neural Network later without breaking your model.

Here are the three specific transformations we need to apply:

1. Log Transformation (For Skewed Data)
Features like Cost and Follower_Count rarely follow a normal distribution; they are usually "right-skewed" (many restaurants are cheap, very few are extremely expensive).

The Transformation: merged_df['cost'] = np.log1p(merged_df['cost'])

Why: Models like Linear Regression or Neural Networks assume data is somewhat normally distributed. A log transformation "squashes" the long tail, reducing the impact of outliers and making the data distribution more Gaussian (Bell curve), which helps the model learn faster.

2. Feature Scaling (Standardization)
In your dataset, Cost might range from 0 to 5000, while Rating ranges from 1 to 5. Without scaling, the model might incorrectly assume that Cost is 1000x more important than Rating just because the numerical values are larger.

The Transformation: StandardScaler (centers the mean at 0 and variance at 1).

Why: It puts all features on an "equal playing field." It ensures the model weights are assigned based on predictive power, not the unit of measurement (e.g., Rupee vs. Star Rating).

3. One-Hot Encoding (For Categorical Data)
Your dataset likely contains categorical columns like Cuisine_Type (Italian, Indian, Chinese) or Location.

The Transformation: pd.get_dummies() or OneHotEncoder.

Why: Machine learning models are mathematical. They cannot perform calculations on the word "Italian." One-hot encoding creates a binary column (0 or 1) for each category, allowing the math to run on numeric inputs

````python
import numpy as np
from sklearn.preprocessing import StandardScaler

# 1. Log Transform skewed features to normalize the distribution
merged_df['cost'] = np.log1p(merged_df['cost'])
merged_df['follower_count'] = np.log1p(merged_df['follower_count'])

# 2. Convert Categorical variables into numbers
# Fixed: Changed 'Cuisine' to 'cuisines' to match your lowercase column names
if 'cuisines' in merged_df.columns:
    merged_df = pd.get_dummies(merged_df, columns=['cuisines'], drop_first=True)

# 3. Scale all numerical features
scaler = StandardScaler()
numerical_cols = ['cost', 'follower_count', 'review_count']
merged_df[numerical_cols] = scaler.fit_transform(merged_df[numerical_cols])

print("Data transformation complete. Features are now normalized and ready.")
````

### 9. Handling Imbalanced Dataset

##### Do you think the dataset is imbalanced? Explain Why.

Yes, restaurant rating datasets are inherently imbalanced, and this is a significant hurdle for any regression model.

Why is the dataset imbalanced?

In a Zomato or Yelp dataset, ratings are almost never uniformly distributed. You will typically see a "Central Tendency Bias":

The "Bell Curve" Skew: Most restaurants cluster in the 3.5 to 4.2 rating range.

The Sparse Tails: Very high ratings (4.8–5.0) are rare (often niche), and very low ratings (1.0–2.0) are also rarer (most restaurants go out of business before consistently hitting these numbers).

````python
import numpy as np
import pandas as pd
from sklearn.utils.class_weight import compute_sample_weight
from sklearn.ensemble import RandomForestRegressor

# 1. Create temporary bins to calculate frequency
# We bin the continuous ratings (e.g., 1-5) into 5 discrete categories
y_binned = pd.cut(y_train, bins=5, labels=False, include_lowest=True)

# 2. Calculate weights automatically
# 'balanced' mode automatically adjusts weights inversely proportional to class frequencies
weights = compute_sample_weight(class_weight='balanced', y=y_binned)

# 3. Train the model using weights
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train, sample_weight=weights)

print("Model trained with sample weighting to prioritize extreme ratings.")
````

##### What technique did you use to handle the imbalance dataset and why? (If needed to be balanced)

Handling the Imbalance (Regression Approach):

Since we are performing Regression (predicting a continuous number) rather than Classification, we cannot use standard techniques like SMOTE (which is designed for categorical classes). Instead, I recommend Sample Weighting.

The Technique: Sample Weighting

We assign higher "importance" (weights) to the observations in the rare tails (1.0 or 5.0) and lower weight to the observations in the middle (the majority). This forces the model to care more about getting the extreme cases right.

#### 2. Cross- Validation & Hyperparameter Tuning

````python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error

# 1. Define the parameter grid
param_grid = {
    'max_depth': [5, 10, 15, 20],
    'min_samples_split': [2, 10, 20],
    'min_samples_leaf': [1, 5, 10]
}

# 2. Initialize and Fit the Algorithm with GridSearchCV
dt_grid = GridSearchCV(DecisionTreeRegressor(random_state=42), param_grid, cv=3, scoring='r2', n_jobs=-1)
dt_grid.fit(X_train, y_train)

# Best parameters
print(f"Best Parameters (Decision Tree): {dt_grid.best_params_}")

# 3. Predict on the model
y_pred_dt = dt_grid.best_estimator_.predict(X_test)

# 4. Evaluate performance
r2_dt = r2_score(y_test, y_pred_dt)
mse_dt = mean_squared_error(y_test, y_pred_dt)

print(f"Tuned Decision Tree R2 Score: {r2_dt:.4f}")
print(f"Tuned Decision Tree MSE: {mse_dt:.4f}")
````
